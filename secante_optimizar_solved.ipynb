```python
def secant_method_s(f, x0, x1, tol=1e-6, max_iter=100):    
    x_prev = x0
    f_prev = f(x_prev)  
    x_curr = x1
    f_curr = f(x_curr)  
    iter_count = 0

    while abs(f_curr) > tol and iter_count < max_iter:
        x_next = x_curr - f_curr * (x_curr - x_prev) / (f_curr - f_prev)

        x_prev, f_prev = x_curr, f_curr
        x_curr, f_curr = x_next, f(x_next)
        iter_count += 1

    return x_curr, iter_count
```

```python
i = 0


def func(x):
    global i
    i += 1
    y = x**3 + 2 * x**2 + 3 * x - 1
    print(f"Llamada i={i}\t x={x:.5f}\t y={y:.2f}")
    return y


secant_method_s(func, x0=2, x1=3)
```
